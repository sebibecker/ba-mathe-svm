## Darstellung der Transformation durch die Funktion $\phi$

### Von 1d nach 2d

```python
plt.grid(alpha=0.5)
plt.scatter(a, np.zeros_like(a), color="b")
plt.scatter(b, np.zeros_like(b), color="r")
plt.savefig("images/1d_2d_1.png")
```

```python
f = lambda x: x**2
```

```python
h = np.arange(-5,5,1)
plt.grid(alpha=0.5)
plt.scatter(a, f(a), color="b")
plt.scatter(b, f(b), color="r")
plt.axhline(y=2.5, color="g", linestyle="--")
plt.savefig("images/1d_2d_2.png")
```

### Von 2d nach 3d

```python
a = np.random.rand(30, 2) - np.array([0.5, 0.5])
angle = np.linspace(0, 2*np.pi, 200)
r = np.random.rand(200) + 2
b = np.array([(r*np.cos(angle), r*np.sin(angle))])
```

```python
plt.grid(alpha=0.5)
plt.xlim(-3.5, 3.5)
plt.ylim(-3.5, 3.5)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.scatter(a[:,0], a[:,1], color="r")
plt.scatter(b[:,0], b[:,1], color="b")
plt.savefig("images/2d_3d_1.png")
```

```python
f = lambda x,y: np.sqrt(2) * x * y
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')
ax.scatter(a[:,0]**2, f(a[:,0], a[:,1]), a[:,1]**2, color="r")
ax.scatter(b[:,0]**2, f(b[:,0], b[:,1]), b[:,1]**2, color="b")
ax.view_init(15, 110, 0)
ax.set_xlim(-0.5, 9)
ax.set_ylim(-0.5, 9)
ax.set_zlim(-0.5, 9)

xx, zz = np.meshgrid(np.linspace(-1,5), np.linspace(-1,3))
h = - zz + 2
ax.plot_surface(h, xx, zz, alpha=0.4, color="g")
plt.savefig("images/2d_3d_2.png")
```

## Beispiel Soft-Margin

```python
angle = np.linspace(0, 2*np.pi, 100)
r = np.random.rand(100)*2
a = np.array([(r*np.cos(angle)+1.1, r*np.sin(angle)+1.1)])
b = np.array([(r*np.cos(angle)-1.1, r*np.sin(angle)-1.1)])

x = np.linspace(-3,3,100)
h = - x
```

```python
plt.grid(alpha=0.5)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.scatter(a[:,0], a[:,1], color="b")
plt.scatter(b[:,0], b[:,1], color="r")
plt.plot(x, h, color="g", linestyle="--")
plt.savefig("images/soft_margin.png")
```

## Darstellung optimale Hyperebene

```python
angle = np.linspace(0, 2*np.pi, 20)
r = np.random.rand(20)
r2 = np.random.rand(20)
a = np.array([(r*np.cos(angle)-2, r*np.sin(angle)+2)])
support_a = np.array([(-2,0), (-1.5, 0.5), (0, 2)])
b = np.array([(r2*np.cos(angle)+2, r2*np.sin(angle)-2)])
support_b = np.array([(0.5,-1.5), (1,-1)])

x = np.linspace(-4,4,100)
h = x
h_1 = x + 2
h_2 = x - 2
```

```python
plt.grid(alpha=0.5)
plt.xlim(-3.3, 3.3)
plt.ylim(-3.3, 3.3)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.plot(x, h, color="g", linestyle="--")
plt.text(0, 0.5, r"$H$")
plt.plot(x, h_1, color="k", linestyle=":")
plt.text(1, 2.7, r"$H_1$")
plt.plot(x, h_2, color="k", linestyle=":")
plt.text(2.5, 1.3, r"$H_{-1}$")
plt.fill_between(x, h_1, h_2, color="0.75", alpha=0.5)
plt.scatter(a[:,0], a[:,1], color="b")
plt.scatter(support_a[:,0], support_a[:,1], color="b", s=60)
plt.scatter(b[:,0], b[:,1], color="r")
plt.scatter(support_b[:,0], support_b[:,1], color="r", s=60)

plt.arrow(x=-2, y=-2, dx=-1, dy=1, linestyle="--", color="c")
plt.arrow(x=-2, y=-2, dx=1, dy=-1, linestyle="--", color="c")
plt.text(-2.8, -1.8, r"$d$")
plt.text(-1.8, -2.8, r"$d$")
plt.savefig("images/optimale_hyperebene.png")
```

## Darstellung m√∂glicher klassifizierender Hyperebenen

```python
angle = np.linspace(0, 2*np.pi, 50)
r = np.random.rand(50)
r2 = np.random.rand(50)
a = np.array([(r*np.cos(angle)-1.5, r*np.sin(angle)+1.5)])
b = np.array([(r2*np.cos(angle)+1.5, r2*np.sin(angle)-1.5)])

x = np.linspace(-4,4,100)
h = x
h1 = 0.5*x + 0.5
h2 = -0.25*x
```

```python
plt.grid(alpha=0.5)
plt.xlim(-3.3, 3.3)
plt.ylim(-3.3, 3.3)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.scatter(a[:,0], a[:,1], color="b")
plt.scatter(b[:,0], b[:,1], color="r")
plt.plot(x, h, color="g", linestyle="--")
plt.plot(x, h1, color="c", linestyle="--")
plt.plot(x, h2, color="m", linestyle="--")
plt.savefig("images/possible_hyperplanes.png")
```
